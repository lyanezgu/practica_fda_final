```python
#eliminar dolar de columna de las últimas 3 y cambiamos los nombres de las columnas
#para seguir un patrón en los nombres


def clean_data(data_df):
    # Preparing a dict with old columns as keys, with new columns as values
    old_cols = data_df.columns.values
    new_cols = {}
    for i in range(len(old_cols)):
        new_cols[old_cols[i]] = old_cols[i].strip().replace(' ', '_').lower()
    # Renaming all columns names via pandas rename() function
    data_df = data_df.rename(columns=new_cols)
    
    # Parsing financial values as floats
    data_df['average_covered_charges'] = data_df['average_covered_charges'].str[1:].astype(float)
    data_df['average_total_payments'] = data_df['average_total_payments'].str[1:].astype(float)
    data_df['average_medicare_payments'] = data_df['average_medicare_payments'].str[1:].astype(float)

    return data_df


```

```python
cleaned_data_df = clean_data(data_df)
```
